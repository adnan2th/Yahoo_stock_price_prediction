We can't assume a uniform nor normal distribution in the plot above, because of that the statistical summary can helps us more, for this we will split the data into 10 chunks and compute mean and variance of each one.



It calculates the mean and standard deviation of the 'High' column in your dataframe df, in groups of 182 rows, and then shows them together in a table.

```python
from statistics import  stdev
```

```python
means = []
std = []

for i in range(0,10):
  means.append(df.High[(i*182):(i*182)+182].mean())
  std.append(stdev(df['High'].iloc[(i*182):(i*182)+182]))
```

This loop runs 10 times (from i = 0 to i = 9).

Each time, it takes a chunk of 182 rows from the 'High' column.

For example:

When i=0: rows 0 to 181

When i=1: rows 182 to 363

When i=2: rows 364 to 545,

```python
pd.concat([pd.DataFrame(means,columns=['mean']),pd.DataFrame(std,columns=['std'])], axis=1)
```

A small std → prices were stable (less variation).

A large std → prices were volatile (more ups and downs).

The mean shows your trend — prices are mostly increasing over time.

The std shows your stability — some periods (like 6, 8, and 9) have high volatility, meaning prices moved a lot in those parts of data.

#Normliaztion



```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['High'] = scaler.fit_transform(df[['High']])
```

#Split Data

```python
df1 = df['High']
```

```python
train_size = int(len(df1)*0.7)
test_size = int(len(df1)*0.15)
valid_size = len(df1) - train_size - test_size

train = df.iloc[0:train_size]
test = df.iloc[train_size:train_size+test_size]
valid = df.iloc[train_size+test_size:len(df1)]

print(train.shape, test.shape, valid.shape)
```

```python
df1
```

```python
train.drop('Date', axis=1, inplace=True)
test.drop('Date', axis=1, inplace=True)
valid.drop('Date', axis=1, inplace=True)
```

#Create Sequence

```python
def create_squence(data, sequence_length):
    x = []
    y = []
    for i in range(len(data)-sequence_length):
        x.append(data[i:(i+sequence_length),0])
        y.append(data[i+sequence_length, 0]) # Modified to select only the first column for y
    return np.array(x),np.array(y)
```

```python
X_train, y_train = create_squence(train.values, 5)
X_test, y_test   = create_squence(test.values, 5)
X_valid, y_valid = create_squence(valid.values, 5)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print(X_valid.shape, y_valid.shape)
```

# reshape



```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
X_valid = X_valid.reshape(X_valid.shape[0], X_valid.shape[1], 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
```

# Define Model

```python
# Model def
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Dropout ,LSTM ,GRU
```

```python
model = Sequential()
model.add(SimpleRNN(100,activation='relu', input_shape=(5,1),return_sequences= True))
model.add(Dropout(0.2))
model.add(SimpleRNN(100,activation='relu', return_sequences= True))
model.add(Dropout(0.2))
model.add(SimpleRNN(100,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
model.summary()
```

#Compile

```python
model.compile(optimizer='RMSprop', loss='mse', metrics=['mean_absolute_error'])
```

#Regularizer

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(

    monitor='val_loss',

    patience=10,

    restore_best_weights=True # Restores the weights from the best epoch
)
```

#Model fit

```python
history = model.fit(X_train, y_train, epochs=50 , validation_data=(X_valid,y_valid),callbacks=[early_stopping])
```

```python
plt.figure(figsize=(18,8))
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
```

```python
plt.figure(figsize=(18,8))
plt.plot(history.history['mean_absolute_error'], label='train')
plt.plot(history.history['val_mean_absolute_error'], label='test')
```

```python
y_pred = model.predict(X_test)
```

#Evaluate

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error ,r2_score
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

#LSTM

```python
model_LSTM = Sequential([
    LSTM(100,activation='relu',input_shape=(5,1),return_sequences= True),
    Dropout(0.2),
    LSTM(100,activation='relu',return_sequences= True),
    Dropout(0.2),
    LSTM(100,activation='relu'),
    Dropout(0.2),
    Dense(1)
])
```

```python
model_LSTM.compile(optimizer='RMSprop', loss='mse', metrics=['mean_absolute_error'])
```

```python
history_LSTM = model_LSTM.fit(X_train, y_train, epochs=50 , validation_data=(X_valid,y_valid),callbacks=[early_stopping])
```

```python
plt.figure(figsize=(18,8))
plt.plot(history_LSTM.history['loss'], label='train')
plt.plot(history_LSTM.history['val_loss'], label='test')
```

```python
plt.figure(figsize=(18,8))
plt.plot(history_LSTM.history['mean_absolute_error'], label='train')
plt.plot(history_LSTM.history['val_mean_absolute_error'], label='test')
```

```python
y_pred_LSTM = model_LSTM.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_pred_LSTM))
print(mean_absolute_error(y_test, y_pred_LSTM))
print(r2_score(y_test, y_pred_LSTM))
```
